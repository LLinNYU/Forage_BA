This Jupyter notebook includes some code to get you started with web scraping. We will use a package called `BeautifulSoup` to collect the data from the web. Once you've collected your data and saved it into a local `.csv` file you should start with your analysis.

### Scraping data from Skytrax

If you visit [https://www.airlinequality.com] you can see that there is a lot of data there. For this task, we are only interested in reviews related to British Airways and the Airline itself.

If you navigate to this link: [https://www.airlinequality.com/airline-reviews/british-airways] you will see this data. Now, we can use `Python` and `BeautifulSoup` to collect all the links to the reviews and then to collect the text data on each of the individual review links.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 30
page_size = 100

reviews = []
table_data = []

# for i in range(1, pages + 1):
for i in range(1, pages + 1):

    print(f"Scraping page {i}")

    # Create URL to collect links from paginated data
    url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"

    # Collect HTML data from this page
    response = requests.get(url)

    # Parse content
    content = response.content
    parsed_content = BeautifulSoup(content, 'html.parser')
    for para in parsed_content.find_all("div", {"class": "text_content"}):
        reviews.append(para.get_text())
    flag = True
    #ChatGPT helped me figure out how to make sense of table structures
    for table in parsed_content.find_all("div", {"class": "review-stats"}):
        individual_stars = []
        table_rows = table.find_all("tr")
#         print(".") #just seeing if it detects the classes at all
        for row in table_rows:
            cells = row.find_all("td")
            feature = cells[0].text
            if cells[1].text == '12345':
                response = str(cells[1]).count('star fill') #YAY
#                 if flag:
#                     print(response)
#                     flag = False
            else: 
                response = cells[1].text
            individual_stars.append((feature, response))
        table_data.append(individual_stars)

    print(f"   ---> {len(reviews)} total reviews")
```

```python
df = pd.DataFrame()
df["reviews"] = reviews
# df["stars"] = table_data
df.head()
```

```python
# Should do some more cleaning
# figure out how may categories to put in the df
categories = {}

#this is for each review
for entry_index in table_data: 
    #not everyone input the same information about features
    for num_features in range(len(entry_index)):
        if entry_index[num_features][0] not in categories:
            categories[entry_index[num_features][0]] = True
        else: 
            next
# there are 13 categories
# print(len(categories))
```

```python
# need to all all the categories into the dataframe
df[list(categories.keys())] = None
df.head()
```

```python
flag = True

#for each person who input star ratings
for entry in range(len(table_data)):
    
    # a row per person
    star_row = table_data[entry]
    
    # for each of the features they mention
    for num_features in range(len(star_row)):
        feature_name = star_row[num_features][0]
        
        if feature_name in categories: 
            if flag:
                print(feature_name)
                flag = False
            df.loc[entry,feature_name] = str(star_row[num_features][1])              
        else: 
            next
df.head()
```

```python
df.to_csv("./BA_reviews.csv")
```
